```python
import pandas as pd

# 1. Load data
file_path = r"C:\Users\USER\Downloads\AMR_Analysis_Project_csv.csv"
df = pd.read_csv(file_path)

# 2. Define the Mapping Dictionary (WHO codes to Anchor Countries)
region_map = {
    'AFR': 'Rwanda', 
    'AMR': 'USA', 
    'EUR': 'Germany', 
    'SEAR': 'India', 
    'EMR': 'Egypt', 
    'WPR': 'Australia'
}

# 3. Apply the mapping to the correct snake_case column
df['mapped_country'] = df['region_or_country'].str.strip().map(region_map)

# 4. Analysis: Average Resistance per Region
# Using 'region_name' (lowercase) or 'Region name' as per your list
avg_resistance = df.groupby('Region name')['numeric_value'].mean().sort_values(ascending=False)

print("✅ Data Mapped and Analyzed Successfully!")
print("\n--- Average Resistance (Risk Level) ---")
print(avg_resistance)

# 5. Save the Cleaned Version for your GitHub Repo
df.to_csv('Cleaned_AMR_Surveillance_Data.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set_theme(style="whitegrid")

# Count reporting density
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Region name', order=df['Region name'].value_counts().index, palette='viridis')

plt.title('Surveillance Data Density: Why Africa Lags Behind', fontsize=15)
plt.xlabel('Number of Surveillance Records Uploaded')
plt.ylabel('WHO Region')
plt.show()
```
